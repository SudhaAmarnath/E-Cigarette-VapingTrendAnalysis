## 4. Pre processing the data set.
The  df_smokefree data set has the catagorial values for example the column MeasureDesription having the row values in string etc.. These values can be encoded by preprosseing the data frame using inbuilt function LabelEncoder. So with this, all the categorial values are assigned a numberic label so that for the next classifications the values can be used instead of the string names. A new data frame df_smokefree_en is created mapping with original df_smokefree. Also in a similar way, following data sets are encoded with label values. 
1. df_smokefree_enc
2. df_preemption_enc
3. df_tax_enc
4. df_riskfactor_enc


```
retainColNamesList = ['LocationDesc', 'LocationAbbr', 'MeasureDesc', 'ProvisionDesc', 'ProvisionGroupDesc', 'ProvisionValue', 'Gender', 'Race', 'Age', 'Education']
retainColRenamedList = ['LocationDescID', 'LocationID', 'MeasureDescID', 'ProvisionDescID', 'ProvisionGroupDescID', 'ProvisionValueID', 'GenderID', 'RaceID', 'AgeID', 'EducationID']

def number_encode_features(df, skipMap=False):
    result = df.copy()
    encoders = {}
    print(result.columns)
    for column in result.columns:
        if result.dtypes[column] == np.object:
            encoders[column] = preprocessing.LabelEncoder()
            try:
              if column in retainColNamesList and skipMap == False:
                idx = retainColNamesList.index(column)
                result[retainColRenamedList[idx]] = encoders[column].fit_transform(result[column].astype(str))
              else:
                result[column] = encoders[column].fit_transform(result[column].astype(str))
            except:
              result[column] = np.NaN
    return result, encoders
```

```
df_smokefree_enc, encodelst = number_encode_features(df_smokefree)
df_preemption_enc, encodelst = number_encode_features(df_preemption)
df_tax_enc, encodelst = number_encode_features(df_tax)
df_riskfactor_enc, encodelst = number_encode_features(df_riskfactor)
dataFramesEncList = [df_smokefree_enc, df_preemption_enc, df_tax_enc, df_riskfactor_enc]
```

## 9. Function to plot the clusters

```
def plot_cluster(cluster_col1, cluster_col2,labels,centroids=[]):
    import matplotlib.pyplot as plt
    plt.figure(figsize=(6, 6))
    scatter = plt.scatter(cluster_col1, cluster_col2, c=labels, cmap='rainbow', s = 10)
    #plt.legend(handles=scatter.legend_elements()[0], labels=labels)
    if len(centroids) > 0:
        plt.scatter(centroids[:, 0], centroids[:, 1], s = 100, c = 'black', label = 'Centroids', marker = 'o')
        plt.legend(loc=(1.05,0.5))
    plt.xlabel(cluster_col1.name)
    plt.ylabel(cluster_col2.name)
    plt.show()
```

## 10. run_cluster_algorithms function is defined.
It is used to calculate the number of clusters N using the Elbow method. This calculated N value is used to run the two clustering algoritms KMEANS, GMM, DBSCAN, BIRCH, MiniBatch KMEANS etc. Also the sklearn metrics ARI, MI, HCV, FM, CH, DB are calculated and compared to choose cluster.

```
def run_cluster_algorithms(cdf, cluster_col1, cluster_col2, elbow=True):
    global N, clusterlabels
    # Using number of clusters N=4 as calculated
    N = 4
    if elbow is True:
        distortions=[]
        K = range(1,10)
        for k in K:
            kmeans = KMeans(n_clusters=k).fit(cdf)
            kmeans.fit(cdf)
            distortions.append(sum(np.min(cdist(cdf, kmeans.cluster_centers_, 'euclidean'), axis=1)) / cdf.shape[0])
        # Plot the elbow
        plt.plot(K, distortions, 'bx-', linewidth=2)
        plt.xlabel('K')
        plt.ylabel('Distortion')
        plt.title('The Elbow Method to find the optimal k')
        plt.show()
        print('Selecting optimum N value =', N)
        print('Found distortion from elbow method for',N, 'clusters =', distortions[N-1])

    metricsdf = pd.DataFrame(index=['KMEANS','GMM', 'DBSCAN', 'MBK', 'BIRCH'], 
                            columns=['ARI','MI','HCV','FM','CH','DB'])

    XX = cdf[cluster_col1]
    YY = cdf[cluster_col1]

    kmeans = KMeans(init='k-means++', n_clusters = N, random_state = 0, max_iter = 500)
    t0 = time()
    kmeans.fit(cdf)
    train_time = time() - t0
    print("\nTime taken to train with KMeans Model using ScikitLearn is %0.2f seconds" % train_time)
    labels = kmeans.labels_
    clusterlabels = labels
    centroids = kmeans.cluster_centers_
    plot_cluster(cdf[cluster_col1], cdf[cluster_col2], labels, centroids)
    metricsdf.loc['KMEANS','ARI'] = metrics.adjusted_rand_score(YY, labels)
    metricsdf.loc['KMEANS','MI']  = metrics.adjusted_mutual_info_score(YY, labels)
    metricsdf.loc['KMEANS','HCV'] = metrics.homogeneity_score(YY, labels)
    metricsdf.loc['KMEANS','FM'] = metrics.fowlkes_mallows_score(YY, labels)
    metricsdf.loc['KMEANS','CH'] = metrics.calinski_harabasz_score(cdf, labels)
    metricsdf.loc['KMEANS','DB'] = metrics.davies_bouldin_score(cdf, labels)


    gmm = mixture.GaussianMixture(N, covariance_type='tied', random_state = 0)
    t0 = time()
    gmm.fit(cdf)
    train_time = time() - t0
    print("\nTime taken to train with GMM Model is %0.2f seconds" % train_time)
    labels = gmm.predict(cdf)
    #calcMetrics('GMM')
    plot_cluster(cdf[cluster_col1], cdf[cluster_col2], labels)
    metricsdf.loc['GMM','ARI'] = metrics.adjusted_rand_score(YY, labels)
    metricsdf.loc['GMM','MI']  = metrics.adjusted_mutual_info_score(YY, labels)
    metricsdf.loc['GMM','HCV'] = metrics.homogeneity_score(YY, labels)
    metricsdf.loc['GMM','FM'] = metrics.fowlkes_mallows_score(YY, labels)
    metricsdf.loc['GMM','CH'] = metrics.calinski_harabasz_score(cdf, labels)
    metricsdf.loc['GMM','DB'] = metrics.davies_bouldin_score(cdf, labels)


    dbscan = DBSCAN(eps=0.1, min_samples = 4)
    t0 = time()
    dbscan.fit(cdf)
    train_time = time() - t0
    print("Time taken to train with DBSCAN Model %0.2f seconds" % train_time)
    labels = dbscan.labels_
    plot_cluster(cdf[cluster_col1], cdf[cluster_col2], labels)
    metricsdf.loc['DBSCAN','ARI'] = metrics.adjusted_rand_score(YY, labels)
    metricsdf.loc['DBSCAN','MI']  = metrics.adjusted_mutual_info_score(YY, labels)
    metricsdf.loc['DBSCAN','HCV'] = metrics.homogeneity_score(YY, labels)
    metricsdf.loc['DBSCAN','FM'] = metrics.fowlkes_mallows_score(YY, labels)
    metricsdf.loc['DBSCAN','CH'] = metrics.calinski_harabasz_score(cdf, labels)
    metricsdf.loc['DBSCAN','DB'] = metrics.davies_bouldin_score(cdf, labels)


    mbk = MiniBatchKMeans(init='k-means++', n_clusters=N, batch_size=100,
                          n_init=10, max_no_improvement=10, verbose=0,
                          random_state=0)
    t0 = time()
    mbk.fit(cdf)
    t_mini_batch = time() - t0
    print("\nTime taken to train with MiniBatchKMeans Model is %0.2f seconds" % t_mini_batch)
    labels = mbk.labels_
    centroids = mbk.cluster_centers_
    plot_cluster(cdf[cluster_col1], cdf[cluster_col2], labels, centroids)
    metricsdf.loc['MBK','ARI'] = metrics.adjusted_rand_score(YY, labels)
    metricsdf.loc['MBK','MI']  = metrics.adjusted_mutual_info_score(YY, labels)
    metricsdf.loc['MBK','HCV'] = metrics.homogeneity_score(YY, labels)
    metricsdf.loc['MBK','FM'] = metrics.fowlkes_mallows_score(YY, labels)
    metricsdf.loc['MBK','CH'] = metrics.calinski_harabasz_score(cdf, labels)
    metricsdf.loc['MBK','DB'] = metrics.davies_bouldin_score(cdf, labels)


    birch = Birch(threshold=1.7, n_clusters=4)

    t0 = time()
    birch.fit(cdf)
    train_time = time() - t0
    print("Time taken to train with Birch Model %0.2f seconds" % train_time)
    labels = birch.labels_
    plot_cluster(cdf[cluster_col1], cdf[cluster_col2], labels, centroids)
    metricsdf.loc['BIRCH','ARI'] = metrics.adjusted_rand_score(YY, labels)
    metricsdf.loc['BIRCH','MI']  = metrics.adjusted_mutual_info_score(YY, labels)
    metricsdf.loc['BIRCH','HCV'] = metrics.homogeneity_score(YY, labels)
    metricsdf.loc['BIRCH','FM'] = metrics.fowlkes_mallows_score(YY, labels)
    metricsdf.loc['BIRCH','CH'] = metrics.calinski_harabasz_score(cdf, labels)
    metricsdf.loc['BIRCH','DB'] = metrics.davies_bouldin_score(cdf, labels)


    '''
    # N is taken from elbow method that was found using k-means
    aggcluster = AgglomerativeClustering(n_clusters=N, affinity='euclidean', linkage='ward')
    t0 = time()
    aggcluster.fit_predict(cdf)
    train_time = time() - t0
    print("Time taken to train with AgglomerativeClustering Model %0.2f seconds" % train_time)
    labels = aggcluster.labels_
    plot_cluster(cdf[cluster_col1], cdf[cluster_col2], labels, centroids)
    metricsdf.loc['AGGLOMERATIVE','ARI'] = metrics.adjusted_rand_score(YY, labels)
    metricsdf.loc['AGGLOMERATIVE','MI']  = metrics.adjusted_mutual_info_score(YY, labels)
    metricsdf.loc['AGGLOMERATIVE','HCV'] = metrics.homogeneity_score(YY, labels)
    metricsdf.loc['AGGLOMERATIVE','FM'] = metrics.fowlkes_mallows_score(YY, labels)
    metricsdf.loc['AGGLOMERATIVE','CH'] = metrics.calinski_harabasz_score(cdf, labels)
    metricsdf.loc['AGGLOMERATIVE','DB'] = metrics.davies_bouldin_score(cdf, labels)


    # N is taken from elbow method that was found using k-means
    specclustering = SpectralClustering(n_clusters=N,assign_labels="discretize",random_state=0)
    t0 = time()
    specclustering.fit(cdf)
    train_time = time() - t0
    print("Time taken to train with SpectralClustering Model %0.2f seconds" % train_time)
    labels = specclustering.labels_
    plot_cluster(cdf[cluster_col1], cdf[cluster_col2], labels, centroids)
    metricsdf.loc['SPECTRAL','ARI'] = metrics.adjusted_rand_score(YY, labels)
    metricsdf.loc['SPECTRAL','MI']  = metrics.adjusted_mutual_info_score(YY, labels)
    metricsdf.loc['SPECTRAL','HCV'] = metrics.homogeneity_score(YY, labels)
    metricsdf.loc['SPECTRAL','FM'] = metrics.fowlkes_mallows_score(YY, labels)
    metricsdf.loc['SPECTRAL','CH'] = metrics.calinski_harabasz_score(cdf, labels)
    metricsdf.loc['SPECTRAL','DB'] = metrics.davies_bouldin_score(cdf, labels)

    '''


    print('Comparing Clusters  from different Algorithms using sklearn metrics')
    display(metricsdf)
```

## 11. Get Top Contets in the Clusters, Visualize Data Clusters

```
def get_top_x_per_cluster(x, col, data, clusters, n):
    top_lists = [dict() for i in range(n)]
    c = 0
    
    for index, row in data.iterrows():
        i = clusters[c]
        c += 1
        if row[col] in top_lists[i]:
            top_lists[i][row[col]] += 1
        else:
             top_lists[i][row[col]] = 1
    
    title = 'Top ' + str(x) + ' ' + str(col) + ' in each cluster' 
    VAR1 = list()
    VAR2 = list()
    print(title)
    for i in range(n):
        print('\nCluster ', i)
        sorted_list = sorted(top_lists[i].items(), key=lambda kv: kv[1])
        sorted_list = list(reversed(sorted_list))
        var1 = list()
        var2 = list()
        for y in range(x):
            print(sorted_list[y])
            var1.append(sorted_list[y][0])
            var2.append(sorted_list[y][1])
        VAR1.append(var1)
        VAR2.append(var2)
    
    plt.figure(figsize=(10, 10))
    plt.suptitle(title, fontsize=20)
    colorlist = ['b', 'r', 'g', 'c', 'm', 'y']
    for i in range(n):
        plt.subplot(221+i)
        plt.bar(VAR1[i],VAR2[i], width = 0.25, color=colorlist[i])
        #plt.barh(VAR1[i], VAR2[i], label = 'Cluster'+str(i))        
        #plt.legend(fontsize=15, loc=(1.05,0.5))
        plt.xticks(rotation=45)
        titlestr = 'Cluster ' + str(i)
        plt.title(titlestr, fontsize=10)
    plt.show()
    
    return

top_x = 3
```

## 12. Function to plot Correlation Map

```
def plot_correlation_map(DF):
    Cor = DF[DF.columns] 
    #Calculate the correlation of the above variables
    cor = Cor.corr() 
    mask = np.zeros_like(cor, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    #Plot the correlation as heat map
    fig, ax = plt.subplots(figsize=(20,20)) 
    sns.heatmap(cor, mask=mask, vmin=-1, vmax=1, center=0,cmap=sns.diverging_palette(20, 220, n=200), annot=True,linewidths=.5, ax=ax)
```

## 28. Function classification_algorithm to run and fit regression algorithms.


```
def classification_algorithm(DF, X_train_cols, y_train_col, clf, classifier):
    print(X_train_cols)
    print(y_train_col)
    X = DF[X_train_cols].iloc[ :, 0:]
    Y = DF[y_train_col]
    X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    calculate_scores(y_test, y_pred, classifier)
    if classifier == 'Decision Tree':
        from IPython.display import Image
        export_graphviz(clf, out_file = "ratingDCTree.dot", 
                                        feature_names =  X_train.columns, 
                                        rounded = True, proportion = False, 
                                        precision = 2, filled = True)
        from subprocess import call
        call(['dot', '-Tpng', 'ratingDCTree.dot', '-o', 'ratingDCTree.png', '-Gdpi=60'])
        # Display in jupyter notebook
        from IPython.display import Image
        display(Image(filename = 'ratingDCTree.png'))
    if classifier == 'Random Forest':
      estimator = clf.estimators_[2]
      # Export as dot file
      export_graphviz(estimator, out_file='rfTree.dot', 
                      feature_names = X_train.columns,
                      rounded = True, proportion = False, 
                      precision = 2, filled = True)
      # Convert to png using system command (requires Graphviz)
      from subprocess import call
      call(['dot', '-Tpng', 'rfTree.dot', '-o', 'rfTree.png', '-Gdpi=60'])
      # Display in jupyter notebook
      from IPython.display import Image
      display(Image(filename = 'rfTree.png'))
```

## 23. Classifier Datasets, Function run_classification_algorithms. Following algoritms are run
1. Logistic Regression
2. Naive Bayes
3. Decision Tree (with tree plot)
4. Random Forest (with tree plot)
5. SVM
6. KNN

```
def run_classification_algorithms(DF, X_train_cols, y_train_col, algorithms='all'):
    global algorithms_performace
    global algorithms_performace_df
    #if algorithms == 'all':
      #algorithms_performace = []
    if algorithms == 'all' or algorithms == 'Logistic Regression':
      classification_algorithm(DF, X_train_cols, y_train_col, LogisticRegression(random_state=100), 'Logistic Regression')
    if algorithms == 'all' or algorithms == 'Naive Bayes':
        classification_algorithm(DF, X_train_cols, y_train_col, GaussianNB(), 'Naive Bayes')
    if algorithms == 'all' or algorithms == 'Decision Tree':
        classification_algorithm(DF, X_train_cols, y_train_col, DecisionTreeClassifier(max_depth=5, random_state=100), 'Decision Tree')
    if algorithms == 'all' or algorithms == 'Random Forest':
        classification_algorithm(DF, X_train_cols, y_train_col, RandomForestClassifier(max_depth=5,n_estimators=3, random_state=100), 'Random Forest')
    if algorithms == 'all' or algorithms == 'SVM':
        classification_algorithm(DF, X_train_cols, y_train_col, SVC(random_state = 100), 'SVM')
    if algorithms == 'all' or algorithms == 'KNN':
      classification_algorithm(DF, X_train_cols, y_train_col, KNeighborsClassifier(n_neighbors = 20, metric = 'minkowski', p = 2), 'KNN')
    algorithms_performace_df = pd.DataFrame(algorithms_performace)
    if algorithms == 'all':    
        algorithms_performace_df.plot(x='algorithm', y='accuracy', kind='bar', color='green')
        display(algorithms_performace_df.head(10))
```

## 46. Data Normalization using Sigmoid Function

```
def normalized_sigmoid(center, width, x):
   s= 1/(1+np.exp(width*(x-center)))
   # normalizing sigmoid function to 0-1
   return (1- 1*(s-min(s))/(max(s)-min(s)))
   
def plot_sigmoid(x): 
    plt.figure(figsize=(10, 8))
    plt.plot(x, normalized_sigmoid(50, .20, x), 'r', label='Lung Injuries as of Oct 8th 2019 (0,102)', linewidth=5)
    plt.grid()
    plt.title('Plot using Sigmoid Function',fontsize=18)
    # place the legend in bottom right of the graph
    plt.legend(loc='lower right')
    # write the Sigmoid formula
    plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=25)
    plt.gca().xaxis.set_major_locator(plt.MultipleLocator(25))
    plt.xlabel('Vaping Related Lung Injury Cases from CDC as of Oct 8th 2019 per State', fontsize=18)
    plt.ylabel('Probability more than 50 Cases', fontsize=18)
    plt.show()

lungInjuryCountNpArray = ndf['LungInjuryCount'].to_numpy()
lungInjuryCountNpArray.sort()
plot_sigmoid(lungInjuryCountNpArray)
```
